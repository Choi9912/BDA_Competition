```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2. 서울시 상권분석서비스(상권-생활인구)
df2 = pd.read_csv('서울시 상권분석서비스(상권-생활인구).csv', encoding="euc-kr") # 사용할 데이터셋
df2.head(2)
```

```python
# 골목상권으로 분류되는 코드만 추출
df2 = df2[df2[' 상권_구분_코드_명']=='골목상권'].reset_index(drop=True)
df2.head(2)
```

```python
# 22년도 데이터만 추출
df2 = df2[(df2['기준 년코드']==2022)].reset_index(drop=True)
df2.head(2)
```

```python
df2['기준 년코드'].value_counts()
```

```python
# 맨 앞의 숫자만 제외하고 끝의 6자리만 남김.
df2['상권_코드'] = df2['상권_코드'].apply(str)
df2['상권_코드'] = df2['상권_코드'].str[1:]
df2['상권_코드'].head()
```

```python
# 다시 정수형으로
df2['상권_코드'] = df2['상권_코드'].apply(int)
```

### 1. 강동구

```python
# '상권_코드'가 111048 이상 111090 이하인 데이터 필터링
강동구_df2 = df2[(df2['상권_코드'] >= 111048) & (df2['상권_코드'] <= 111090)]
강동구_df2
```

```python
강동구_df2['시군구_코드'] = '강동구' # 시군구 컬럼 추가
```

```python
# 가만... 생각해보니 태블로에서 시각화할 거면 이대로 파일 뽑으면 되지 않나?
강동구_df2.to_csv('강동구_인구.csv')
```

```python
# 필터링된 데이터의 컬럼들 합산
total_value_강동 = 강동구_df2.sum()
total_value_강동
```

```python
df = pd.DataFrame(total_value_강동).transpose()
df
```

### 2. 강서구

```python
# 강서구 '상권_코드' 범위는 110629 이상 110679 이하인 데이터 필터링
강서구_df2 = df2[(df2['상권_코드'] >= 110629) & (df2['상권_코드'] <= 110679)]
강서구_df2['시군구_코드'] = '강서구' # 시군구 컬럼 추가
```

### 3. 관악구

```python
# 관악구 '상권_코드' 범위는 110853 이상 110902 이하인 데이터 필터링
관악구_df2 = df2[(df2['상권_코드'] >= 110853) & (df2['상권_코드'] <= 110902)]
관악구_df2['시군구_코드'] = '관악구' # 시군구 컬럼 추가
```

### 4. 중랑구

```python
# 중랑구 '상권_코드' 범위는 110233 이상 110279 이하인 데이터 필터링
중랑구_df2 = df2[(df2['상권_코드'] >= 110233) & (df2['상권_코드'] <= 110279)]
중랑구_df2['시군구_코드'] = '중랑구' # 시군구 컬럼 추가
```

### 5. 광진구

```python
# 광진구 '상권_코드' 범위는 110139 이상 110180 이하인 데이터 필터링
광진구_df2 = df2[(df2['상권_코드'] >= 110139) & (df2['상권_코드'] <= 110180)]
광진구_df2['시군구_코드'] = '광진구' # 시군구 컬럼 추가
```

### 6. 병합

```python
# 데이터프레임들을 병합
result = pd.concat([강동구_df2, 강서구_df2, 관악구_df2, 중랑구_df2, 광진구_df2], ignore_index=True)
result.head()
```

```python
result.shape
```

```python
# '시군구_코드' 컬럼 이름을 '시군구_코드_명'으로 변경
result.rename(columns={'시군구_코드': '시군구_코드_명'}, inplace=True)
result.head(2)
```

```python
result['시군구_코드_명'].value_counts()
```

```python
# csv 파일로 내보내기
result.to_csv('2. 5개_구_유동인구.csv')
```
